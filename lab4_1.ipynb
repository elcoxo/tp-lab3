## Функция f = p * Tr(q * vT) 

Создадим квадратичные матрицы M и N из 85 элементов

```python
# Размер матрицы
size = 85

# Создаем матрицу, заполненную 2 по всему массиву
M = 2 * np.ones((size, size), dtype=int)

# Заполняем главную диагональ числом арифметической прогрессией с шагом от 3 до 171 с шагом 3
np.fill_diagonal(M, np.arange(3, 173, 2))

print("Matrix M\n", M)

# Создаем матрицу, заполненную 4 по всему массиву
N = 4 * np.ones((size, size), dtype=int)
# Заполняем каждую нечетную строку числом 5 после каждого второго элемента
N[0::2, 2::3] = 5
# Заполняем каждую четную строку числом 7 после каждого элемента
N[1::2, 1::2] = 7

print("Matrix N\n", N)
```

Вектор нормального распределения с математическим ожиданием a=2 и дисперсией σ2=4

```python
mean = 2 # математическое ожидание
sigma_squared = 4 # дисперсия 
std_dev = np.sqrt(sigma_squared) # стандартное отклонение

# Генерируем массив чисел, распределенных нормально
p = np.random.normal(loc=mean, scale=std_dev, size=(85,))

print('Vector p\n', p)
```

Транспонированный вектор q

```python
# Выполняем поэлементное умножение нужных строк из матриц N и M с изменением каждого элемента j на 1
q = N[24][:] * M[71][:]

# Транспонирование вектора
qT = q.reshape(-1, 1)

print(qT)
```

Транспонированный вектор v

```python
i = np.arange(85)
v = M[84-i] + 2*N[i]

# Транспонирование вектора
vT = v.reshape(-1, 1)

print(vT)
```

След произведения векторов q и vT

```python
trace_matrix = np.trace(q*vT)
print(trace_matrix)
```

Значение функции f = p * Tr(q * vT) 

```python
f = p * trace_matrix
print(f)
```

## Функция g = || qT * N ||_1

```python
#L1-норма
l1_norm = np.linalg.norm(qT*N, ord=1)
print(l1_norm)
```
